# CREATING A DICTIONARY 

THAT ATTACHES IMAGES TO WQI TARGET VALUE:

```python
def get_dict_from_csv(path):
    directory = os.listdir(path=path)
    directory = [x for x in directory if '.jpg' in x]
    out = dict()
    for label in directory:
#         print(label)
        try:
            out[label] = df6[df6['Longitude']==float(label.replace('.jpg', ''))]['WQI'].values[0]
        except:
            continue
    return out
```

# CREATING A GENERATOR 

THAT FINDS DICT, READS IN IMAGES, STANDARDIZES, RESHAPES AND ASSIGNS BATCHES.

```python
def image_gen(batch_size, y_dict, path):
    directory = os.listdir(path=path)
    Xout = []
    yout = []
    counter = 0
    for filename in it.cycle(directory):
        if '.jpg' in filename:
            try:
                yout.append(y_dict[filename])
                img = cv2.imread(path + filename)
                img = img/255
                img.reshape(1,400,640,3)
                Xout.append(img)
                counter += 1
                if counter % batch_size == 0:
                    yield (np.array(Xout), np.array(yout))
                    Xout = []
                    yout= []
            except:
                continue
```

```python
#set up model
model = Sequential()

model.add(Convolution2D(filters = 6, kernel_size = 7, activation='relu', input_shape=(400,640,3)))
model.add(MaxPooling2D(pool_size=(4,4)))
model.add(Convolution2D(12, kernel_size = 5, activation='relu'))
model.add(MaxPooling2D(pool_size=(3,3)))
model.add(Convolution2D(16, kernel_size = 3, activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Convolution2D(6, kernel_size = 3, activation='relu'))
# model.add(MaxPooling2D(pool_size=(2,2)))
# model.add(Convolution2D(8, kernel_size = 3, activation='relu'))
# model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

from keras import backend as K

def sensitivity(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    return true_positives / (possible_positives + K.epsilon())

def specificity(y_true, y_pred):
    true_negatives = K.sum(K.round(K.clip((1-y_true) * (1-y_pred), 0, 1)))
    possible_negatives = K.sum(K.round(K.clip(1-y_true, 0, 1)))
    return true_negatives / (possible_negatives + K.epsilon())
model.compile('adam', loss='binary_crossentropy', metrics=[sensitivity, specificity, 'accuracy'])

```

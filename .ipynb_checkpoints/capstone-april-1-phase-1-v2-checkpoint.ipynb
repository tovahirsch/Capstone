#### Restrict now the number of painters

Let us restrict the number of painters in the training and test sets separately and then choose only the rows in the rest set which contains painters from the training set:

```python
train['artist'].value_counts().head()
test['artist'].value_counts().head()
```

```python
def threshold(df,minimum):
    artists_count = df['artist'].value_counts()
    artists_count = artists_count[artists_count > minimum]
    artists = artists_count.index.tolist()
    return  df[df['artist'].isin(artists)]
```

```python
train_new = threshold(train,100)
test_new = threshold(test,100)

train_new['artist'].nunique()
test_new['artist'].nunique()
```

#### The function `subfolder_maker` does the following:
- The `for` loops over the images inside the list `source_images` that is either `source_images_train` or `source_images_test` 
- It first opens each image in `os.path.join(PATHTRAIN,image_filename)`, where the image name `image_filename` is in the list of names `source_images` (either `source_images_train` or `source_images_test`), using the `Image` library from `pillow` (which is a python image library)
- The list `original_images` contains the names of training images in `trainsetnew` by 'Pablo Picasso' paintings or the names of Picasso's paintings in the `testsetnew`. 
- The list `original_images_not` contains the names of training images in `trainsetnew` not by 'Pablo Picasso' paintings or the names of paintings **not by** Picasso in the `testsetnew`. 
- The function `img.save( )` will save the files in the subfolder of Picasso and non-Picasso paintings

```python
train_new['new_filename'].tolist()[0:2]
```

```python
source_images_train[0:5]
```

```python
def subfolder_maker(source_images,PATH,PATHDUMP,
                    PATHDUMPNOT,original_images,
                    original_images_not):
    for image_filename in tqdm(source_images):          # source_images_train = ['1.jpg', '10.jpg', '100.jpg',...]
        img = Image.open(os.path.join(PATH, 
                                      image_filename))  # PATH contains all training or testing set
        if image_filename in original_images:           # original_images contains all
            img.save(os.path.join(PATHDUMP, 
                                  image_filename))      # PATHDUMP contains training set with Picasso paintings
        elif image_filename in original_images_not:
            img.save(os.path.join(PATHDUMPNOT, 
                                  image_filename))      # PATHDUMP contains training set with NO Picasso paintings       
```

```python
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K


# dimensions of our images.
img_width, img_height = 256, 256

train_data_dir = '/Users/marcotavora/capstone/trainsetpicasso'
validation_data_dir = '/Users/marcotavora/capstone/testsetpicasso'
train_data_dir
validation_data_dir
nb_train_samples = len(train_picasso['new_filename'].tolist())+len(train_not_picasso['new_filename'].tolist())
nb_validation_samples = len(test_picasso['new_filename'].tolist())+len(test_not_picasso['new_filename'].tolist())
epochs = 5
batch_size = 16

if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

# this is the augmentation configuration we will use for training
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)
```

```python
PATHDUMPPICASSOTRAIN
PATHDUMPNOTPICASSOTRAIN
PATHDUMPPICASSOTEST
PATHDUMPNOTPICASSOTEST
```

```python
test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

history = model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)

model.save_weights('first_try.h5')
```
